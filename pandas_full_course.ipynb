# Handle missing data: fillna, dropna, interpolate, replace

fillna method

```python
df = pd.read_csv('weather_data5.csv',parse_dates=['day'])
df.set_index('day',inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({'temperature':0,'windspeed':0,'event': 'no event'})
new_df
```

```python
new_df = df.fillna(method='ffill') # Forward fill - copy from next cell
new_df
```

```python
new_df = df.fillna(method='bfill') # Before fill - copy from previous cell
new_df
```

```python
new_df = df.fillna(method='ffill',limit=1) # Forward fill, but copy for only once
new_df
```

Interpolate method

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method='time') # Interpolation but more accurate as it considers date
new_df
```

dropna method

```python
new_df = df.dropna() # Drops rows with na values
new_df
```

```python
new_df = df.dropna(how='all') # Drops rows which has all na values
new_df
```

```python
new_df = df.dropna(thresh=1) # Keep rows which has at least 1 valid value
new_df
```

```python
dt = pd.date_range('01-01-2017','01-11-2017')
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```

# Merge Dataframes

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando","baltimore"],
    "temperature": [21,14,35,32],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando","san fancisco"],
    "humidity": [65,68,75,71],
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city') # Same as inner join in MySQL
df3
```

indicator flag

```python
df3 = pd.merge(df1,df2,on='city',how='outer',indicator=True)
df3
```

suffixes

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city')
df3
```

```python
df3 = pd.merge(df1,df2,on='city',suffixes=('_left','_right'))
df3
```
